```python
import plotly.graph_objects as go
import numpy as np
```

```python
t = np.linspace(0,10,100)
y = np.sin(t)
fig = go.Figure(data = go.Scatter(x = t, y=y, mode="lines"))
fig.show()
```

```python
t = np.linspace(0,10,100)
y = np.sin(t)
fig = go.Figure(data = go.Scatter(x = t, y=y, mode="markers"))
fig.show()
```

```python
t = np.linspace(0,10,100)
y = np.sin(t)
fig = go.Figure(data = go.Scatter(x = t, y=y, mode="markers+lines"))
fig.show()
```

```python
N = 100
x = np.linspace(0,1,N)
random_y0 = np.random.randn(N) + 5
random_y1 = np.random.randn(N)
random_y2 = np.random.randn(N) - 5
```

```python
fig = go.Figure()

fig.add_traces(go.Scatter(x=x,y=random_y0, mode = "markers", name = "markers"))

fig.add_traces(go.Scatter(x=x,y=random_y1, mode = "lines", name = "lines"))

fig.add_traces(go.Scatter(x=x,y=random_y2, mode = "markers+lines", name = "markers+lines"))

fig.show()
```

```python
import plotly.graph_objects as go
fig = go.Figure(data = go.Scatter(
    x = [1,2,3,4],
    y=[13,12,11,10],
    mode = 'markers',
    marker = dict(size=[50,70,70,50], color = [0,1,2,3])
))
fig.show()
#tamanho e cor
```

```python
import plotly.graph_objects as go
fig = go.Figure(data = go.Scatter(
    x = [1,2,3,4],
    y=[13,12,11,10],
    mode = 'markers',
    marker = dict(size=[90,80,70,60], color = [0,1,2,3]),
    hovertemplate="R$ %{y} - %{marker.size}"
))
fig.show()
```
